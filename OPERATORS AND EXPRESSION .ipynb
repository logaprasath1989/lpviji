operators are special symbols that designate that some sort of computation should be performed. The values that an operator acts on are called operands.

```python
#Arithmetic operations
a=4
b=3
print("\na=",a)
print("\nb=",b)
#+Unary Positive
print("\n+Unary Positive\n")
print(+a)

#- Unary negative
print("\n-Unary negative\n")
print(-b)

#addition 
print("\nbinary addition\n")
print("sum of a and b:\n",a+b)

#subraction 
print("\nbinary subraction\n")
print("bsubtracted from a:\n",a-b)

#Multiplication
print("\nMultiplication\n")
print("Product of a and b\n",a*b)

#Division
print("\n Division\n")
print("Gives Quotient(float number ) when a is divided by b\n",a/b,type(a/b))

#Modulo 
print("\n Modulo\n")
print("Gives Remainder when a is divided by b\n",a%b,type(a%b))

#floor or interger division
print("\nFloor or interger division\n")
print("Gives Quotient(rounded-whole number ) when a is divided by b\n",a//b,type(a//b))    
      
# Exponentiation
print("\nExponentiation\n")
print("a raised to the power ofb twice\n",a**b)
print("a raised to the power ofb thrice\n",a**b)

```

```python
#Comparison Operators

#The == operator compares the value or equality of two objects-object are same 
print("Comparison Operators",end="\n*")
#Equal to(==)
print("Equal to(==)")
print("\nTrue if the value of a is equal to the value of b,otherwise False\n")
print("20 == 20:",20 == 20)
print("10 == 20:",10 == 20)
print("301==301",301==301)
a=[10,20,3]
b=a.copy()
print("Is objects of a&b is a==b:\n",a==b)
print("id of a",id(a))
print("id of b",id(b))
print("objects of a and b are equal,but does not point to same object")
print(end="\n*")
#Not equal to(!=)
print("Not equal to(!=)")
print("\nTrue if a is not equal to b,otherwise False\n")
print("10!= 20:",10 != 20,type(10 != 20))
print(end="\n*")
#Less than(<)
print("Less than(<)")
print("\nTrue if a is less than b,otherwise False\n")
print("10<20:",10 < 20)
print(end="\n*")
#Less than or equal to(<=)
print("Less than or equal to(<=)")
print("\nTrue if a is less than or equal to b,otherwise False\n")
print("10<=20:",10 <= 20)
print(end="\n*")
#Greater than(>)
print("Greater than(>)")
print("\nTrue if a is greater than b,otherwise False\n")
print("10>20:",10 > 20)
print(end="\n*")
#Greater than or equal to(>=)
print("Greater than or equal to(>=)")
print("\nTrue if a is greater than or equal to b,otherwise False\n")
print("10>=20:",10 >= 20)
```

```python
#Bitwise Operators
#Bitwise operators treat operands as sequences of binary digits and operate on them bit by bit. The following operators are supported:
x = 10 #(0000 1010 in binary) 
y = 4 #(0000 0100 in binary)
print("Bitwise operator\n",end="\n\n*")
#bitwiseAND(&)
print("bitwiseAND(&)\n")
print("position of bits both are 1,then 1,otherwise 0\n")
print("bitwiseAND(&)\nx & y =",x & y,end="\n\n*" )

#bitwise OR(|)
print("bitwise OR(|)\n")
print("position of bits either is 1,then 1,otherwise 0\n")
print("bitwise OR(|)\nx | y=",x | y,end="\n\n*")

#bitwise XOR (exclusive OR)(^)
print("bitwise XOR (exclusive OR)(^)\n")
print("1 if the bits in the operands are different, 0 if they are the same\n")
print("bitwise XOR (exclusive OR)(^)\nx ^ y = ",x ^ y  ,end="\n\n*")

#bitwise negation(~)
print("bitwise negation(~)\n")
print("1 if 0, 0 if 1")
print("bitwise negation(~)\n~x =",~x ,end="\n\n*")

#Shift right n places(>>)
print("Shift right n places(>>)\n")
print("Each bit is shifted right n places\n")
print("Shift right(>>)\nx >> 2 =",x >> 2 ,end="\n\n*")

#Shift left n places(<<)
print("Shift left n places(<<)\n")
print("Each bit is shifted left n places\n")
print("Shift left(<<)\nx >> 2 =",x << 2 ,end="\n")
```
