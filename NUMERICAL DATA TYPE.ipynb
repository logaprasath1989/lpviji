NUMERICAL DATA TYPES 

These are created by numeric literals and returned as results by arithmetic operators and arithmetic built-in functions. Numeric objects are immutable; once created their value never changes. 

There are three distinct numeric types: integers, floating point numbers, and complex numbers

integers-values with positive and negative whole numbers,output values-int,represented by  -int()

float- real numbers ,output values-floating point number ,represented by -float()

complex- contains real and complex numbers output values-set of complex numbers , represented by -complex()

```python
#variable assigment 
x=2.5
y=-2.5
# everything in python is represented  by objects or by relations between objects ,Every object has an identity, a type and a value
print("value of x=",x) # object value 
print("type of x =",x,type(x)) #object type here is int ,here python is dynamic language were no typecaste is requried 
print("identity of x=",id(x)) #object identity  ,here were python is an high level language 
print("type of y=",y,type(y))


#maximum range of float decimal 
float_max=1.000000000000005
print(float_max)
float_max1=1.00000000000000105
print(float_max1)


#how to separate big float like  billion 
billion =1_000_000_000.000# its used for readbility 
print(billion)

#e nottaion 

#postive e 
e_postive=2e15
print(e_postive)
e_ve_range=2e16

#negative e 
print(e_ve_range)
e_neg=2e-4
print(e_neg)

#infinity e#1.8e30
e_infy=1.7e404

print(e_infy)



#float function 
int_num=15
str_num="16"

print(float(int_num))
print(float(str_num))


#float formating 
flt=500000000125.2563

print(f"{flt:.1f}")

print(f"{flt:.2f}")

print(f"{flt:,.3f}")


print("%f is an billion "%flt)
```
